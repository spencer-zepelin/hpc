```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import struct
```

```python
n = 300
```

```python
f = open("f0.bin", "rb")
x = np.fromfile(f, dtype=np.dtype('f8'))
x.shape = (n, n)
ax = sns.heatmap(x, xticklabels=50, yticklabels=50, vmin=0, vmax=1).set_title("At Initialization")
sns.plt.show()
```

```python
sns.plt.clf()
f.close()

f = open("fb0.bin", "rb")
x = np.fromfile(f, dtype=np.dtype('f8'))
x.shape = (n, n)
ax1 = sns.heatmap(x, xticklabels=50, yticklabels=50, vmin=0, vmax=1).set_title("At Initialization")
sns.plt.show()
```

```python
sns.plt.clf()
f.close()

f = open("f1.bin", "rb")
x = np.fromfile(f, dtype=np.dtype('f8'))
x.shape = (n, n)
ax1 = sns.heatmap(x, xticklabels=50, yticklabels=50).set_title("Midway")
sns.plt.show()
```

```python
sns.plt.clf()
f.close()

f = open('f1.txt', 'r')
x = np.loadtxt(f)
ax = sns.heatmap(x, xticklabels=50, yticklabels=50).set_title("Midway")

sns.plt.show()
```
